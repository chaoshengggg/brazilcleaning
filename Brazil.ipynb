#### The code above have examined the order_status and the missing values
 Notice that order_status that are delivered has no missing values at all but for order_status ("invoiced, "shipped",  "approved", "created", "canceled", "unavailable", "shipped") contains missing values.
#### Order_status for "invoiced "always have missing values on the order_delivered_customer_date column
#### Order_status for "shipped " always have missing values on the order_delivered_customer_date column
#### Order_status for "processing" always have missing values on the order_delivered_carrier_date AND order_delivered_customer_date column
#### Order_status for "unavailable" always have missing values on the order_delivered_carrier_date AND order_delivered_customer_date column
#### Order_status for "canceled"  have missing values on the order_approved_at, order_delivered_carrier_date AND order_delivered_customer_date column
#### Order_status for "created"  always have missing values on the order_approved_at, order_delivered_carrier_date AND order_delivered_customer_date column
#### Order_status for "approved" always have missing values on the order_delivered_carrier_date AND order_delivered_customer_date column
 This could because of cancellation of orders, lost track of parcels or the orders are forgotten to be updated to the system. Thus, if we were to handle the missing the values by using prediction method or interpolation it would not be accurate.
Dropping the rows would be the ideal method for this scenario




```python
### Drop rows that has missing values
orders.dropna(inplace=True)
```
